```python
# import of standard Python libraries for data analysis
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# I would like to see all columns of dataframes
pd.set_option('display.max_columns', None)
```

```python
# reading dataframe with snow and meteo variables
snow_meteo = pd.read_csv("snow_meteo.csv")
snow_meteo.head()
```

```python
# checking data types and no nulls in the variables
snow_meteo.info()
```

```python
# variablemassif_num is in fact integer, so I will changed from float to int
snow_meteo.massif_num = snow_meteo.massif_num.astype(int)
```

```python
# in final dataset we need also names of massifs as categorical values
# therefore I need to read dataframe with massif names
shapefile = pd.read_csv("shapefile.csv")
```

```python
# viewing dataframe
shapefile
```

```python
# selecting only needed variables
shapefile = shapefile[["massif_num", "nom_reduit"]]
```

```python
# merge of snow and meteo variables with massif names
snow_meteo = snow_meteo.merge(shapefile, left_on="massif_num", right_on="massif_num")
```

```python
# viewing the result of merge
snow_meteo
```

```python
# to final dataframe I also need to add number of avalanche accidents
# reading Excel file with this information
file = '.\\accidents_massifs.xlsx'
df = pd.read_excel(file)
df.head()
```

```python
# selecting only data from Als and needed variables
df_accidents = df[df["region"]=="alps"]
df_accidents = df_accidents[["date", "massif","massif_unique"]]
df_accidents.date = df_accidents.date.astype(str)

# viewing the result
df_accidents.head()
```

```python
# verifying no null values are in dataset
df_accidents.info()
```

```python
# merge of avalanche accidents with main dataset
snow_meteo = snow_meteo.merge(df_accidents, how="left", left_on=["day","nom_reduit"], right_on=["date", "massif_unique"])
```

```python
snow_meteo.info()
```

```python
# NaN values are days without avalanche accidents, we can transform them all to nulls
snow_meteo = snow_meteo.fillna(0)
```

```python
# checking the result
snow_meteo
```

```python
# verifying avalanche accidents were correctly merged
(snow_meteo[snow_meteo["date"]!=0]).sample(20)
```

```python
# removing columns with doubled information from dataset
snow_meteo = snow_meteo.drop(columns=["massif"])
```

```python
# changing date to 1 to annote that avalanche accident has happened
snow_meteo.date = np.where(snow_meteo["date"]!=0, 1, 0)
```

```python
# checking the result
snow_meteo.date.value_counts()
```

```python
# to final dataframe I also need to add number of avalanche events
# reading file with this information
aval_alps = pd.read_csv("aval_alps.csv")
aval_alps.head()
```

```python
# checking new dataframe for null values
aval_alps.info()
```

```python
# merge of avalanche events with main dataset
data = snow_meteo.merge(aval_alps, how="left", left_on=["day","massif_unique"], right_on=["day_of_event", "massif_name"])
```

```python
# NaN values are days without avalanche events, we can transform them all to nulls
data = data.fillna(0)
```

```python
# verifying avalanche events were correctly merged
(data[data["day_of_event"]!=0]).sample(20)
```

```python
# changing day of event to 1 to annote that avalanche event has happened
data.day_of_event = np.where(data["day_of_event"]!=0, 1, 0)
```

```python
# viewing all columns to decide which are redundant for further analysis
data.columns
```

```python
# removing redundant columns
data = data.drop(columns=['massif_unique', 'country', 'massif_name'])
```

```python
# renaming some columns (date and day_of_event)
data.columns = ['elevation', 'massif_num', 'lon', 'lat', 'temp_soil_0.005_m',
       'temp_soil_0.08_m', 'liquid_water_in_soil', 'frozen_water_in_soil',
       'risk_index', 'aval_type', 'whiteness_albedo', 'net_radiation',
       'drainage', 'runoff', 'snow_melting_rate', 'rainfall_rate',
       'surface_temperature', 'surface_snow_amount', 'thickness_of_snowfall',
       'snow_thickness_1D', 'snow_thickness_3D', 'snow_thickness_5D',
       'snow_thickness_7D', 'snow_water_1D', 'snow_water_3D', 'snow_water_5D',
       'snow_water_7D', 'penetration_ram_resistance',
       'thickness_of_wet_snow_top_of_snowpack',
       'thickness_of_frozen_snow_top_of_snowpack', 'acccidental_risk_index',
       'day', 'surface_air_pressure_mean', 'near_surface_humidity_mean',
       'relative_humidity_mean', 'freezing_level_altitude_mean',
       'rain_snow_transition_altitude_mean', 'air_temp_max', 'wind_speed_max',
       'snowfall_rate_max', 'nebulosity_max', 'air_temp_min', 'nom_reduit', 'aval_accident',
        'aval_event']
```

```python
# viewinf the dataframe after changes
data.head()
```

```python
# removing rows with too low elevation for avalanche to likely occur
data = data[data["elevation"]>1499.0]
```

```python
# display size of dataset
data.shape
```

```python
# checking variables in th dataset
data.info()
```

```python
# selecting better name for variable with massif names
data.rename(columns={'nom_reduit': 'massif_name'}, inplace=True)
```

```python
# saving final datatframe into csv file for later use
final_dataset = data.to_csv(r'.\\final_data.csv', index = False)
```
